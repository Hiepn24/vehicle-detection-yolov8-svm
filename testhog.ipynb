```python
import matplotlib.pyplot as plt
from skimage import color, exposure, io
from skimage.feature import hog
```

```python
image_path = 'data/Bus/ggjvgp_UWZN_jpg.rf.82d6400fe639aa343c1d9b9b4c271421.jpg' 
image = io.imread(image_path)
gray_image = color.rgb2gray(image)
```

```python
features, hog_image = hog(gray_image,
                          orientations=9,
                          pixels_per_cell=(8, 8),
                          cells_per_block=(2, 2),
                          visualize=True,
                          block_norm='L2-Hys')
```

```python
# Tăng độ tương phản ảnh HOG
hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 10))
```

```python
# Hiển thị ảnh gốc và ảnh HOG
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6), dpi=100)

ax1.imshow(gray_image, cmap=plt.cm.gray)
ax1.set_title('Ảnh gốc (xám)')
ax1.axis('off')

ax2.imshow(hog_image_rescaled, cmap=plt.cm.gray)
ax2.set_title('Ảnh sau HOG')
ax2.axis('off')

plt.show()
```
